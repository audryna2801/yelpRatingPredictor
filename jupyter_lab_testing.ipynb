```python
resto_lists_csv="all_restos_lst.csv"
scraped_dir="scraped_data/"

with open(resto_lists_csv, newline='') as f:
        reader = csv.reader(f)
        resto_list = list(reader)

for i in range(2,19):
    resto_csv = scraped_dir + str(i) + ".csv"
    print("crawling this url" + " " + resto_list[i][0])
    with open(resto_csv, "w") as f:
        csvwriter = csv.writer(f)
        csvwriter.writerow(["Rating", "Text"])
        crawl_and_scrape.crawl_resto(resto_list[i][0].strip("'"), csvwriter)
```
